```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('./train.csv', index_col="PassengerId")
```

```python


columns_of_X = [u'Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
```

```python
def data_clean(data):
    columns_of_X = [u'Pclass','Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
    X = data[columns_of_X]
    X['Sex'].replace({'male':1, 'female':0}, inplace=True)
    X['Age'].fillna(0, inplace=True) # 将Age中的NaN转换为0
    X['Embarked'].replace({
        'S': 0,
        'C': 1,
        'Q': 2,
        np.nan: 3
    }, inplace=True)
    X.fillna(0,inplace=True)
    X.applymap(lambda x: int(x))
    return X

train_X = data_clean(train_data)
```

```python
train_Y = train_data['Survived']
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(train_X,train_Y)
```

```python
test_data = pd.read_csv('./test.csv', index_col='PassengerId')
test_X = data_clean(test_data)
```

```python
test_X
```

```python
predict_y = clf.predict(test_X)
```

```python
predict_y
```

```python
result = pd.DataFrame(data=predict_y, index=range(892, 1310))
```

```python
result.to_csv('my_result.csv', index_label=['PassengerId','Survived'])
```
